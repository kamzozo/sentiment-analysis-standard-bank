```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

df = pd.read_csv('sentiment_analysis.csv')

def get_sentiment(comment):
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(comment)['compound']
    if sentiment > 0.2:
        return 'Positive'
    elif sentiment < -0.2:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['comment'].apply(get_sentiment)

# Create pie chart
counts = df['sentiment'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Sentiment Analysis')
plt.show()

# Extract most positive and very negative comments
df_sorted = df.sort_values(by='sentiment')
most_positive = df_sorted[df_sorted['sentiment'] == 'Positive']['comment'].tail(3).tolist()
very_negative = df_sorted[df_sorted['sentiment'] == 'Negative']['comment'].head(3).tolist()
neutral = df_sorted[df_sorted['sentiment'] == 'Neutral']['comment'].head(3).tolist()

print('Top three positive comments:')
for comment in most_positive:
    print(comment)

print('\nTop three negative comments:')
for comment in very_negative:
    print(comment)

print('\nTop three neutral comments:')
for comment in neutral:
    print(comment)

# Save updated DataFrame to CSV
df.to_csv('sentiment_analysis_one.csv', index=False)

```
